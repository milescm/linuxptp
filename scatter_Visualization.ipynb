```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./f_csv/ptpTime5min_7.csv')
```

```python
df.drop(['Unnamed: 0'],axis='columns',inplace=True)
```

```python
# Delete s0, s1 state
# Because the difference in values is too large
# Check the data visualiaztion from the s2 state.

delete_rownum = [0,1]
df = df.drop(delete_rownum)
df = df.reset_index(drop=True)
```

```python
# Add abs master offset column to df_deleteErrdf, df, errdf

absMasterOffset = []

absMasterOffset = df['Master Offset'].abs()

df['Abs Master Offset'] = absMasterOffset
```

```python
# Scatter : offset during 5min

plt.figure(figsize=(15,10))
plt.scatter(x='UTC', y='Master Offset', data=df)
plt.xticks(np.arange(0,310,step=10), ["{:0<2d}".format(x) for x in np.arange(0,310, step=10)],
          fontsize=10,
          rotation=45)
plt.ylim(-6000,8000)
plt.xlabel('Running Time (sec)', labelpad=30, fontsize=15)
plt.ylabel('Offset', labelpad=30, fontsize=15)
plt.title('Offset during 5min', pad=30, fontsize=20)
```
